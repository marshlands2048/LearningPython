# Import CSV file


```python
# Read second column only
import pandas as pd
df = pd.read_csv('data.csv', usecols = ['Joined date'], low_memory = True)
```

```python
df.dtypes
```

```python
df.rename(columns={'Joined date': 'joined'}, inplace=True)
```

```python
df.joined
```

```python
df['date']=df['joined'].str[:15]
```

```python
df.date
```

```python
# Change the data to Date Format of Pandas 
# https://towardsdatascience.com/working-with-datetime-in-pandas-dataframe-663f7af6c587
df['joined'] = pd.to_datetime(df['date'], format='%a %b %d %Y' )
```

```python
df.joined
```

```python
df.dtypes
```

```python
aa = pd.value_counts(df.joined).to_frame().reset_index()
ab = aa.sort_values(by='index')
print(ab)
```

```python
ac = ab.reset_index()
ac.to_csv("export.csv")
```
